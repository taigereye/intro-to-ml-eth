```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

%matplotlib inline
```

```python
train_data = pd.read_csv("data/train.csv")
train_data = train_data.drop(columns='Id')
train_data.head()
```

```python
X_train = train_data.iloc[:,1:]
X_train.head()
```

```python
y_train = train_data.iloc[:,0:1]
y_train.head()
```

```python
for i in range(1,6):
    original = "x" + str(i)
    # Squared terms.
    col_sq = "x" + str(i) + "_sq"
    X_train[col_sq] = np.square(X_train[original])
    # Exp terms.
    col_exp = "x" + str(i) + "_exp"
    X_train[col_exp] = np.exp(X_train[original])
    # Cosine terms.
    col_cos = "x" + str(i) + "_cos"
    X_train[col_cos] = np.cos(X_train[original])
    
X_train["const"] = np.ones(len(X_train))
X_train.head()
```

```python
reg = LinearRegression(fit_intercept=False).fit(X_train, y_train)
print("Coefficients: ", reg.coef_[0])
```

```python
weights_submit = pd.Series(reg.coef_[0])
final_submit = weights_submit.to_csv("submit.csv", index=False)
```
