```python
y_train = train_data.iloc[:,0:1]
y_train.head()
```

```python
X_test = pd.read_hdf("data/test.h5", "test")
X_test.head()
```

```python
scaler = StandardScaler()  
scaler.fit(X_train)  
X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)  
```

```python
def count_classes(predictions):
    counts = []
    for c in range(0, 5):
        count = np.sum([1 if i == c else 0 for i in predictions])
        counts.append(count)
    for c in range(0, 5):
        print(c, ": ", counts[c])
```

```python
net = MLPClassifier(solver='lbfgs', alpha=1e-6, hidden_layer_sizes=(50,30,20,15), random_state=1)
net.fit(X_train, y_train)                         
pred_sk = net.predict(X_test)

count_classes(pred_sk)
```

```python
nodes = 20
input_dim = X_train.shape[1]
output_dim = X_train.shape[0]
hidden_dim = int(input_dim/nodes)

class MultiClassClassifierModule(nn.Module):
    def __init__(
            self,
            input_dim=input_dim,
            hidden_dim=hidden_dim,
            output_dim=output_dim,
            dropout=0.5,
    ):
        super(MultiClassClassifierModule, self).__init__()
        self.dropout = nn.Dropout(dropout)

        self.hidden = nn.Linear(input_dim, hidden_dim)
        self.output = nn.Linear(hidden_dim, output_dim)

    def forward(self, X, **kwargs):
        X = F.relu(self.hidden(X))
        X = self.dropout(X)
        X = F.softmax(self.output(X), dim=-1)
        return X
    
net = NeuralNetClassifier(
    MultiClassClassifierModule,
    max_epochs=20,
    verbose=0
)

clf = LabelPowerset(classifier=net, require_dense=[True,True])
clf.fit(X_train.astype(numpy.float32),y_train)
pred_lp = clf.predict(X_test.astype(numpy.float32))
```
