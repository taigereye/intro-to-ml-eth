```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
train_data = pd.read_csv("data/train.csv")
train_data = train_data.drop(columns=['Id'])
train_data.head()
```

```python
X_train = train_data.iloc[:,1:]
X_train.head()
```

```python
y_train = train_data.iloc[:,0:1]
y_train.head()
```

```python
X_test = pd.read_csv("data/test.csv")
X_test = X_test.drop(columns=['Id'])
X_test.head()
```

```python
reg = LinearRegression().fit(X_train, y_train)
print("Coefficients: ", reg.coef_[0])
print("Intercept: ", reg.intercept_[0])
```

```python
y_pred = pd.DataFrame({'y': reg.predict(X_test).flatten()})
y_pred.head()
```

```python
y_actual = pd.DataFrame({'y': np.mean(X_test, axis=1)})
y_actual.head() 
```

```python
print("R^2: ", reg.score(X_test, y_pred))
print("MSE: ", mean_squared_error(y_actual, y_pred))
```

```python
y_submit = y_pred
y_submit['Id'] = y_submit.index
y_submit = y_submit[['Id', 'y']]
final_submit = y_submit.to_csv("submit.csv", index=False)
```
