```python
def ridge_reg_MSE(l, X_cvin, y_cvin, X_cvout, y_cvout):
    reg = Ridge(alpha=(1 / l))
    reg.fit(X_cvin, y_cvin)
    y_pred = reg.predict(X_cvout)
    return mean_squared_error(y_cvout, y_pred)
```
